```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('HR_comma_sep.csv')
df.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
#df.plot(x='left', y='satisfaction_level', kind = 'box')
df.boxplot(column = 'satisfaction_level', by = 'left',grid =False)
plt.title('Satisfaction level')
plt.suptitle("")
plt.show()
```

```python
#df.plot(x='left', y='satisfaction_level', kind = 'box')
df.boxplot(column = 'last_evaluation', by = 'left',grid =False)
plt.title('Last evaluation score')
plt.suptitle("")
plt.show()
```

```python
df.boxplot(column = 'time_spend_company', by = 'left',grid =False)
plt.title('Time spent in the company')
plt.suptitle("")
plt.show()
```

```python
df.groupby('left').hist(figsize=(12,16))
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df,figsize = (12,8))
plt.show()

#plt.show()
```

```python
current_palette = sns.color_palette()

plt.scatter(np.asarray(df['satisfaction_level']), np.asarray(df['last_evaluation']), s =4,cmap = 'jet', c =np.asarray(df['left']))
plt.show()
```
